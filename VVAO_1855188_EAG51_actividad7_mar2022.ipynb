```python
#librerias utilizadas
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#fijar el tamaño de las graficas
plt.rcParams['figure.figsize'] = [10, 7.5]
```

## Proceso de medias móviles MA(q)
$Y_t = \varepsilon_t - \theta_1 \varepsilon_{t-1} +\theta_2 \varepsilon_{t-2} +\theta_3 \varepsilon_{t-3}+.....+\theta_q \varepsilon_{t-q}$ 

MA(2) con $ \theta_1=0.8 ,  \theta_2=0.8,$ 

$Y_t = \varepsilon_t + 0.8 \varepsilon_{t-1}+ 0.8 \varepsilon_{t-2}$

```python
np.random.seed(29)
ma2 = np.array([1, 0.8, 0.8])
ar2 = np.array([1])
proceso_MA2 = ArmaProcess(ar2, ma2).generate_sample(1000)
```

```python
#graficamos
plt.plot(proceso_MA2)
plt.title(r'Proceso MA(2) $Y_t = \varepsilon_t + 0.8 \varepsilon_{t-1}+ 0.8 \varepsilon_{t-2}$')
plt.show()
```

```python
#Prueba de raices unitarias
result_adf = adfuller(np.cumsum(proceso_MA2))
print('p-value: %.2f' % result_adf[1])
```

```python
#graficamos función de autocorrelación
plot_acf(proceso_MA2, zero=False, lags=30)
plt.title(r'Autocorrelación MA(2) $Y_t = \varepsilon_t + 0.8 \varepsilon_{t-1}+ 0.8 \varepsilon_{t-2}$ ')
plt.show()
```

```python
#graficamos función de autocorrelación parcial
plot_pacf(proceso_MA2, zero=False, lags=30)
plt.title(r'Autocorrelación parcial MA(2) $Y_t = \varepsilon_t + 0.8 \varepsilon_{t-1}+ 0.8 \varepsilon_{t-2}$')
plt.show()  
```

## Proceso Autorregresivo AR(p) 
$Y_t =  \phi_1 Y_{t-1} +\phi_2 Y_{t-2} +\phi_3 Y_{t-3}+.....+\phi_p Y_{t-p}+ \varepsilon_t$ 

 AR(2) con $ \phi_1=0.5$ ,  $ \phi_2=0.1$,

$Y_t = 0.5 Y_{t-1}+0.1 Y_{t-2}+\varepsilon_t$ 

```python
np.random.seed(12)
ma2 = np.array([1])
ar2 = np.array([1, -0.5, -0.1])
proceso_AR2 = ArmaProcess(ar2, ma2).generate_sample(1000)
```

```python
#graficamos
plt.plot(proceso_AR2)
plt.title(r'Proceso AR(2) $Y_t = 0.5 Y_{t-1}+0.1 Y_{t-2}+\varepsilon_t$')
plt.show()
```

```python
#Prueba de raices unitarias
result_adf = adfuller(np.cumsum(proceso_AR2))
print('p-value: %.2f' % result_adf[1])
```

```python
#graficamos función de autocorrelación
plot_acf(proceso_AR2, zero=False, lags=30)
plt.title(r'Autocorrelación AR(2) $Y_t = 0.5 Y_{t-1}+ 0.1 Y_{t-2}+\varepsilon_t$')
plt.show()
```

```python
#graficamos función de autocorrelación parcial
plot_pacf(proceso_AR2, zero=False, lags=30)
plt.title(r'Autocorrelación parcial AR(2) $Y_t = 0.5 Y_{t-1}+ 0.1 Y_{t-2}+\varepsilon_t$')
plt.show()  
```

## Proceso Autorregresivo ARMA(p,q)
